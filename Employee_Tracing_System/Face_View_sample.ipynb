```python
# Parameters
name = "Piyush"
Email = "infopiyush2004@gmail.com"
Organization = "SMVDU"
EmployeeID = "23bcs061"
```

```python
import cv2
import pickle
import numpy as np
import os
video=cv2.VideoCapture(0)
facedetect=cv2.CascadeClassifier('data/haarcascade_frontalface_default.xml')

faces_data=[]
i=0

while True:
    ret,frame=video.read()
    gray=cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)
    faces=facedetect.detectMultiScale(gray, 1.3 ,5)
    for (x,y,w,h) in faces:
        crop_img=frame[y:y+h, x:x+w, :]
        resized_img=cv2.resize(crop_img, (50,50))
        if len(faces_data)<=50 and i%10==0:
            faces_data.append(resized_img)
        i=i+1
        cv2.putText(frame, str(len(faces_data)), (50,50), cv2.FONT_HERSHEY_COMPLEX, 1, (0,0,255), 1)
        cv2.rectangle(frame, (x,y), (x+w, y+h), (0,0,255), 1)
    cv2.imshow("Frame",frame)
    k=cv2.waitKey(1)
    if k==ord('q') or len(faces_data)==50:
        break
video.release()
cv2.destroyAllWindows()

faces_data=np.asarray(faces_data)
faces_data=faces_data.reshape(50, -1)

if faces_data.shape == (50, 7500):
    if 'names.pkl' not in os.listdir('data/'):
        names=[name]*50
        with open('data/names.pkl', 'wb') as f:
            pickle.dump(names, f)
    else:
        with open('data/names.pkl', 'rb') as f:
            names=pickle.load(f)
        names=names+[name]*50
        with open('data/names.pkl', 'wb') as f:
            pickle.dump(names, f)
    
    if 'faces_data.pkl' not in os.listdir('data/'):
        with open('data/faces_data.pkl', 'wb') as f:
            pickle.dump(faces_data, f)
    else:
        with open('data/faces_data.pkl', 'rb') as f:
            faces=pickle.load(f)
        faces=np.append(faces, faces_data, axis=0)
        with open('data/faces_data.pkl', 'wb') as f:
            pickle.dump(faces, f)
    
    if 'Emails.pkl' not in os.listdir('data/'):
        Emails=[Email]*50
        with open('data/Emails.pkl', 'wb') as f:
            pickle.dump(Emails, f)
    else:
        with open('data/Emails.pkl', 'rb') as f:
            Emails=pickle.load(f)
        Emails=Emails+[Email]*50
        with open('data/Emails.pkl', 'wb') as f:
            pickle.dump(Emails, f)
            
    
    if 'Organizations.pkl' not in os.listdir('data/'):
        Organizations=[Organization]*50
        with open('data/Organizations.pkl', 'wb') as f:
            pickle.dump(Organizations, f)
    else:
        with open('data/Organizations.pkl', 'rb') as f:
            Organizations=pickle.load(f)
        Organizations=Organizations+[Organization]*50
        with open('data/Organizations.pkl', 'wb') as f:
            pickle.dump(Organizations, f)
    
    
    if 'EmployeeIDs.pkl' not in os.listdir('data/'):
        EmployeeIDs=[EmployeeID]*50
        with open('data/EmployeeIDs.pkl', 'wb') as f:
            pickle.dump(EmployeeIDs, f)
    else:
        with open('data/EmployeeIDs.pkl', 'rb') as f:
            EmployeeIDs=pickle.load(f)
        EmployeeIDs=EmployeeIDs+[EmployeeID]*50
        with open('data/EmployeeIDs.pkl', 'wb') as f:
            pickle.dump(EmployeeIDs, f)

else:
    print("Feature vector shape mismatch; cannot append!")
```
