```python
import tkinter as tk
from tkinter import ttk
from tkinter import messagebox
import papermill as pm
import subprocess
from datetime import datetime

date = datetime.now().strftime("%d-%m-%Y")

def run_faceAdd_file():
    name = entry_name.get()
    Email = entry_email.get()
    Organization =Entry_Organization.get()
    EmployeeID = Entry_EmployeeID.get()
    
    if not name or not Email or not Organization or not EmployeeID:
             messagebox.showwarning("Input Error", "Please fill above details...")
             return
    
    confirm =messagebox.askyesno("Running", f"Starting Face collection process...")
    if confirm:  
        try:       
            pm.execute_notebook( "Employee_sample.ipynb", "Employee_sample.ipynb",
                                parameters={'name': name, 'Email': Email,'Organization': Organization, 'EmployeeID': EmployeeID})
      
            messagebox.showinfo("Success", "Face added successfully!")
            
        except Exception as e:
            messagebox.showerror("Error", f"Execution failed:\n{e}")
        

def run_recognition_file():
    confirm =messagebox.askyesno("Running", "Starting Attendance process...")
    if confirm:    
        try:
            pm.execute_notebook("Employee_recognition.ipynb", "Employee_recognition.ipynb")
            
            messagebox.showinfo("Success", "Attendance Taken successfully!")
            
        except Exception as e:
            messagebox.showerror("Error", f"Execution failed:\n{e}")

def run_Employee_Tracing_file():      
    confirm =messagebox.askyesno("Running", "Starting Tracing process...")
    if confirm:    
        try:
            pm.execute_notebook("Employee_Tracing.ipynb", "Employee_Tracing.ipynb")
            
            messagebox.showinfo("Success", "Tracing Done successfully!")
            
        except Exception as e:
            messagebox.showerror("Error", f"Execution failed:\n{e}")


def run_OpenCsv_file():
    csv_file = f"Attendance/Attendance_{date}.csv"
    try:
        subprocess.Popen(['start', csv_file], shell=True)
    except Exception as e:
        print(f"Error opening file: {e}")

def run_Open_Absence_Log_file():
    csv_file = f"Logs/Log_{date}/absence_log_{date}.csv"
    try:
        subprocess.Popen(['start', csv_file], shell=True)
    except Exception as e:
        print(f"Error opening file: {e}")

def run_Open_Detection_Log_file():
    csv_file = f"Logs/Log_{date}/detection_log_{date}.csv"
    try:
        subprocess.Popen(['start', csv_file], shell=True)
    except Exception as e:
        print(f"Error opening file: {e}")
            

def reset_entry(): 
    entry_name.delete(0, tk.END)                
    entry_email.delete(0, tk.END)
    Entry_Organization.delete(0, tk.END)
    Entry_Organization.set("Select Organization")
    Entry_EmployeeID.delete(0, tk.END)



def exit_app():
    confirm = messagebox.askyesno("Exit", "Are you sure you want to exit?")
    if confirm:
        root.destroy()

Organizations = ["SMVDU","GOOGLE","MICROSOFT","APPLE",
               "FACEBOOKS","NETFLIX","AMAZON","MINTRA"]
        
# GUI Setup
root = tk.Tk()
root.title("Employee Attendance & tracing Management Dashboard")
root.geometry("520x590")
root.configure(bg="#d4f739")

# Main headline label
headline = tk.Label(root, text="Facial Recognition Based Employee Monitoring System", font=("Helvetica", 15, "bold"),
                    bg="#d4f739",fg="#800000")
headline.pack(pady=2)

tk.Label(root, text="Enter Your Name:", fg="#076834", font=("Helvetica", 15, "bold"), bg="#d4f739").pack(pady=5)
entry_name = tk.Entry(root, bg="#fdffb6", width=35)
entry_name.pack(ipady=1)

tk.Label(root, text="Enter Your Email:", fg="#076834", font=("Helvetica", 15, "bold"), bg="#d4f739").pack(pady=5)
entry_email = tk.Entry(root, bg="#fdffb6", width=35)
entry_email.pack(ipady=1)

# New row for Organization and Entry Numbers
row_frame = tk.Frame(root, bg="#d4f739")
row_frame.pack(pady=5)
tk.Label(row_frame, text="Organization:", fg="#076834", font=("Helvetica", 15, "bold"), bg="#d4f739").grid(row=0, column=0, padx=10)
selected_Org = tk.StringVar()
Entry_Organization = ttk.Combobox(row_frame,textvariable=selected_Org,values=Organizations,state="readonly",width=25)
Entry_Organization.set("Select Organization")
Entry_Organization.grid(row=1, column=0, padx=5)
tk.Label(row_frame, text="Employee ID:", fg="#076834", font=("Helvetica", 15, "bold"), bg="#d4f739").grid(row=0, column=1, padx=10)
Entry_EmployeeID = tk.Entry(row_frame, bg="#FFFFFF", width=27)
Entry_EmployeeID.grid(row=1, column=1, padx=10,ipady=1)

#Buttons
tk.Button(root, text="Reset", bg="#FFD700", width=12, height=1, font=("Helvetica", 9, "bold"), command=reset_entry).pack(pady=10)

tk.Button(root, text="Employee Registration", bg="#1b960b", fg="white", width=57, height=1, font=("Helvetica", 10, "bold"), 
          command=run_faceAdd_file).pack(pady=5)


btn_row = tk.Frame(root, bg="#d4f739")
btn_row.pack(pady=10)

tk.Button(btn_row, text="Take Attendance",bg="#2196F3", fg="white",width=24, height=2,font=("Helvetica", 10, "bold"),
          command=run_recognition_file).pack(side=tk.LEFT, padx=6)

tk.Button(
    btn_row, text="Start Monitoring",bg="#F0515C", fg="white",width=24, height=2,font=("Helvetica", 10, "bold"),
    command=run_Employee_Tracing_file).pack(side=tk.LEFT, padx=6)


row2 = tk.Frame(root, bg="#d4f739")
row2.pack(pady=10)

tk.Button(row2, text="View Detection Logs",bg="#736b73", fg="white",width=17, height=2,font=("Helvetica", 10, "bold"),
          command=run_Open_Detection_Log_file).pack(side=tk.LEFT, padx=6)

tk.Button(row2, text="View Absence Logs",bg="#AB7E82", fg="white",width=17, height=2,font=("Helvetica", 10, "bold"),
          command=run_Open_Absence_Log_file).pack(side=tk.LEFT, padx=6)

tk.Button(root, text="View Attendance", bg="#736E28", fg="white", width=28, height=2, font=("Helvetica", 10, "bold"), 
          command=run_OpenCsv_file).pack(pady=10)

tk.Button(root, text="Exit", bg="#f44336", fg="white", width=18, height=1, font=("Helvetica", 10, "bold"), 
          command=exit_app).pack(pady=10)

headline = tk.Label(root, text="Thankyou!!", font=("Helvetica", 15, "bold"),
                    bg="#d4f739",fg="#D44CCF")
headline.pack(pady=2)

root.mainloop()
```
